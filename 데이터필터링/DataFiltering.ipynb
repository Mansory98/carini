```python
# csv -> df
df = pd.read_csv('./chatbot_project/data/car_data.csv', delimiter=',', encoding='cp949')
```

```python
# df 기본정보
display(df.shape)
display(df.info())
display(df[['가격(최저가)','가격(최고가)']].describe())
display(df.head())
display(df.columns)
```

```python
data = {"가격":"5000만원 이상 1억원 미만", "크기":"소형", "연료":"가솔린"}
```

```python
# 가격 필터링

if data["가격"]=="5000만원 미만":
    df_price = df[df["가격(최저가)"]<5000]
elif data["가격"]=="5000만원 이상 1억원 미만":
    df_price = df[(df["가격(최저가)"]>=5000) & (df["가격(최저가)"]<10000)]
elif data["가격"]=="1억원 이상":
    df_price = df[df["가격(최저가)"]>=10000]
else:
    df_price = df

# 크기 필터링
if data["크기"] != "선택안함":
    df_size = df_price[df_price["크기"].str.split(" ").str[0] == data["크기"]]
else:
    df_size = df_price

# 연료 필터링

if data["연료"] != "선택안함":
    df_fuel = df_size[df_size["연료"].str.strip().str.split(",").apply(lambda x: data["연료"] in x)]
else:
    df_fuel = df_size

display(df_fuel)
```

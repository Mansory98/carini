```python
# car_data_onlynum.csv 파일의 데이터를 DB에 넣는 코드입니다. 컬럼명은 car_fuel(연료) , car_eff(연비) 입니다.
import MySQLdb
import csv

# MySQL 연결 정보
host = 'localhost'
user = 'root'
password = '12345'
database = 'carini'
port = 3306

try:
    # 데이터베이스에 연결
    conn = MySQLdb.connect(host=host, user=user, password=password, database=database, port=port)

    # 커서 생성
    cursor = conn.cursor()

    # csv 파일의 경로
    csv_file_path = "./data/car_data_onlynum.csv"

    # csv 데이터 넣기
    with open(csv_file_path, 'r', encoding='cp949') as file:
        csv_data = csv.reader(file)
        next(csv_data)  # 헤더 스킵
        for row in csv_data:
            query = "INSERT INTO car_info (car_name, car_min_price, car_max_price, car_size, car_fuel, car_eff , car_img) VALUES (%s, %s, %s, %s, %s, %s , %s)"
            values = (row[0], row[1], row[2], row[3], row[4], row[5] , row[6]) #위 쿼리문의 insert의 values 값에 대응하여 넣기
            cursor.execute(query, values)

    # 변경사항 커밋
    conn.commit()

except Exception as e:
    # 예외 발생 시 롤백
    conn.rollback()
    # 예외 내용 출력
    print("Error:", e)

finally:
    # 연결 종료
    conn.close()
```

```python
# 자동차 평점데이터 csv파일의 데이터를 db로 넣는 코드입니다.
import MySQLdb
import csv

# MySQL 연결 정보
host = 'localhost'
user = 'root'
password = '12345'
database = 'carini'
port = 3306

try:
    # 데이터베이스에 연결
    conn = MySQLdb.connect(host=host, user=user, password=password, database=database, port=port)

    # 커서 생성
    cursor = conn.cursor()

    # csv 파일의 경로
    csv_file_path = "./data/항목별 점수(도형).csv"

    # csv 데이터 넣기
    with open(csv_file_path, 'r', encoding='cp949') as file:
        csv_data = csv.reader(file)
        next(csv_data)  # 헤더 스킵
        for row in csv_data:
            query = "INSERT INTO car_score (name, price, habit, quality, desgin, eff , avg) VALUES (%s, %s, %s, %s, %s, %s , %s)"
            values = (row[0], row[1], row[2], row[3], row[4], row[5] , row[6]) #위 쿼리문의 insert의 values 값에 대응하여 넣기
            cursor.execute(query, values)

    # 변경사항 커밋
    conn.commit()

except Exception as e:
    # 예외 발생 시 롤백
    conn.rollback()
    # 예외 내용 출력
    print("Error:", e)

finally:
    # 연결 종료
    conn.close()
```

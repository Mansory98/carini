```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import pyperclip
import time
```

```python
# 자동차 페이지 넘기기 

## 네이버 자동차 페이지 접속  *정상
driver = webdriver.Chrome()
driver.get('https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=0&ie=utf8&query=%EC%9E%90%EB%8F%99%EC%B0%A8') # 로그인하기위한 사이트 주소
time.sleep(0)  # 3초간 지연(대기시간)


names=[]
sizes=[]
prices=[]
oils=[]
years=[]
for index in range(1,53):
    for cnt in range(1, 9):
        print('#main_pack > div.sc_new.cs_common_module.case_list.color_5._list_car > div.cm_content_wrap > div.cm_content_area > div > div.horizon_box_image.type_rel._car_panel_wrapper > div:nth-child('+str(index)+') > div:nth-child('+str(cnt)+') > div.info_area > strong > a')
        nextPage_btn = driver.find_element(By.CSS_SELECTOR, '#main_pack > div.sc_new.cs_common_module.case_list.color_5._list_car > div.cm_content_wrap > div.cm_content_area > div > div.horizon_box_image.type_rel._car_panel_wrapper > div:nth-child('+str(index)+') > div:nth-child('+str(cnt)+') > div.info_area > strong > a')
        
        nextPage_btn.click()
        time.sleep(3)
        
        # 자동차 상세정보(class : "more_link") 클릭 
        login_btn = driver.find_element(By.CLASS_NAME, 'more_link')
        login_btn.click()
        time.sleep(3)
        
        # 이름 가져오기
        car_name = driver.find_elements(By.CLASS_NAME, '_text') 

        # 크기 가져오기
        car_size = driver.find_elements(By.CSS_SELECTOR, '#main_pack > div.sc_new.cs_common_module.case_normal.color_5._cs_car_single > div:nth-child(3) > div > div.title_area.type_keep._title_area > div.sub_title > span:nth-child(1)')

        # 자동차 정보 가져오기 
        car_infos = driver.find_elements(By.CLASS_NAME, 'info_group')
        
        
        for name in car_name:
            print(name.text)
            names.append(name.text)
            break
        for size in car_size:
            print(size.text)
            sizes.append(size.text)
            break
        
        for info in car_infos:
            if '가격' in info.text:
                car_price=info.text.split("\n")[1]
                print(car_price)
                prices.append(car_price)
            elif '연료' in info.text:
                car_oil=info.text.split("\n")[1]
                print(car_oil)
                oils.append(car_oil)
            elif '연비' in info.text:
                car_year=info.text.split("\n")[1]
                print(car_year)
                years.append(car_year)
      
        driver.back()
        time.sleep(5)

        driver.back()
        time.sleep(5)
    # 자동차 페이지 넘김 클릭 
    login_btn1 = driver.find_element(By.CSS_SELECTOR, '#main_pack > div.sc_new.cs_common_module.case_list.color_5._list_car > div.cm_content_wrap > div.cm_content_area > div > div.cm_paging_area.no_margin > div > a.pg_next.on')
    login_btn1.click()
    time.sleep(0)

```

```python

import pandas as pd

data={
    "이름" : names,
    "크기" : sizes,
    "가격" : prices,
    "연료" : oils,
    "연비" : years
    
}

df=pd.DataFrame(data)

df.to_csv("./car_info.csv",index=False,encoding="utf-8-sig")
```
